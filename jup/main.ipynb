```python
import pandas as pd 
```

```python
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(10, size=(5,6)), columns=list("ABCDEF"))
```

```python
df.head()
```

```python
df.query("A > 4")
```

```python
df.query("A > 4 and E == 5")
```

```python
df.query("A > 2 or F > 4")
```

```python
df.query("not (A > 2 or F > 4)")
```

```python
df.query("A*2 < 5")
```

```python
df.F = -1 * df.F
df.query("abs(F) > 6")
```

```python
df.query("sqrt(A) > 2")
```

```python
df["G"] = ["John", "Jane", "Ashley", "Luke", "Michael"]
df.query("G > 'Jo'")
```

```python
df["Date"] = pd.date_range(start="2021-04-05", periods=5, freq="M")
```

```python
df.head()
```

```python
df.query("Date.dt.month > 6")
```

```python
df.query("A + B < C")
```
